```python
import json
```

```python
# Load model_list
with open('./model_types.json', 'r') as f:
    model_list = json.load(f)
```

```python
print(model_list)
```

```python
# Expand the dictionary
new_model_list = {}
for model_type in model_list:
    for model_architectures in model_list[model_type]:
        new_model_list[model_architectures] = model_list[model_type][model_architectures]
```

```python
# Save new_model_list as json
with open('./model_archs.json', 'w') as f:
    json.dump(new_model_list, f)
```
