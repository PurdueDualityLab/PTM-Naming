```python
from openai import OpenAI
import json
from dotenv import load_dotenv
import os
import torch
import pandas as pd

load_dotenv()
client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
)

naming_category = ['A', 'I', 'IX', 'O', 'T']
```

# Regular Data format: one naming category per message

## GPTs Prompt- 100 datapoints

### temperature = 0.2

```python
import constant2
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
file_path = "data/PTM_naming_train.jsonl"

y_true = []
y_pred = []

with open(file_path, "r") as file:
  lines = file.readlines()
for line in lines:
  obj = json.loads(line)
  for content in obj.values():
    input = content[1]
    chatlog = []
    chatlog.append({"role" : "system", "content" : constant2.BACKGROUND})
    chatlog.append(input)
    completion = client.chat.completions.create(
      model="gpt-4-0125-preview",
      messages=chatlog,
      temperature=0.2
    )
    output = completion.choices[0].message.content
    print(input['content'])
    
    print("Prediction: ", output)
    print("Actual value: ", content[2]['content'])
    y_true.append(content[2]['content'])
    y_pred.append(output)

conf_matrix = confusion_matrix(y_true, y_pred, labels=naming_category)
print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=naming_category))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import json

file_path = "data/FINAL_data/FINAL_lang_revised_test.jsonl"

y_true = []
y_pred = []
thres = 90

with open(file_path, "r") as file:
    lines = file.readlines()

for line in lines:
    obj = json.loads(line)
    messages = obj["messages"]
    # system_content = messages[0]["content"]
    user_contents = messages[0]["content"]
    # print(f'Length: {len(user_contents)}')
    assistant_contents = messages[1]["content"]
    
    for user_content, assistant_content in zip(user_contents, assistant_contents):
        chatlog = []
        chatlog.append({"role": "system", "content": constant12_6.BACKGROUND})
        chatlog.append({"role": "user", "content": user_content})
        completion = client.chat.completions.create(
            model="gpt-4-0125-preview",
            messages=chatlog,
            temperature=0.2,
            logprobs=True,
            top_logprobs=2
        )
        output = completion.choices[0].message.content

        top_two_logprobs = completion.choices[0].logprobs.content[0].top_logprobs
        prob = np.round(np.exp(completion.choices[0].logprobs.content[0].logprob)*100,2)
        if prob > thres:
            y_true.append(assistant_content)
            y_pred.append(output)

            print(user_content)
            print("Prediction: ", output)
            print("Actual value: ", assistant_content)
            for i, logprob in enumerate(top_two_logprobs, start=1):
                print(f"Output token {i}: {logprob.token}, logprobs: {logprob.logprob}, linear prob: {np.round(np.exp(logprob.logprob)*100,2)}")
            print("\n")
# Assuming `naming_category` holds the list of all possible categories
conf_matrix = confusion_matrix(y_true, y_pred, labels=naming_category)
print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=naming_category))
```
